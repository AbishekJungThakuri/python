```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ["c","c++","Java","python"]
ex = [0.4,0.0,0.0,0.0]

c = ['r','b','g','y']
# colors=c startangle=90

plt.pie(x,labels=y,explode=ex, autopct="%0.1f%%",shadow=True,radius=1,labeldistance=1.1,textprops={"fontsize":15})

plt.legend(loc=2)
plt.show()
```
