```python
import matplotlib.pyplot as plt
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
plt.scatter(day,no,color='y')
plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Number")
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
no2 = [3,2,4,5,1,6,2]
# colors = ['y', 'r', 'b', 'g', 'm', 'b', 'g']  

colors = [10,49,30,29,56,20,30]

sizes = [100,20,40,50,30,10,70]

plt.scatter(day, no, c=colors, s = sizes,marker="*")
plt.scatter(day, no2, c=colors, s = sizes,marker="*")

t = plt.colorbar()
t.set_label("ColorBar")

plt.title("Scatter Plot")
plt.xlabel("Day")
plt.ylabel("Number")
plt.show()
```
