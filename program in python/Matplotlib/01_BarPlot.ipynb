```python
import matplotlib.pyplot as plt
```

```python
x = ["Python","c","c++","java"]
y = [80,57,70,77]
plt.xlabel("Programming languages",fontsize=15)
plt.ylabel("Number of users",fontsize=15)
plt.title("List of languages")
# c = ['y','b','g','m']
plt.bar(x,y,width=0.5,color='r',label="Popularity" )
plt.legend()
plt.show()
```

```python
# Multiple bar graph
import numpy as np

x = ["Python","c","c++","java"]
y = [80,57,70,77]
z = [20,30,40,50]

width=0.3
p=np.arange(len(x))
p1 = [j+width for j in p]
plt.xlabel("Programming languages",fontsize=15)
plt.ylabel("Number of users",fontsize=15)
plt.title("List of languages")

plt.bar(p,y,width,color='r',label="Popularity" )
plt.bar(p1,z,width,color='y',label="Popularity1" )

plt.xticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```

```python
x = ["Python","c","c++","java"]
y = [80,57,70,77]
z = [20,30,40,50]

plt.xlabel("Programming languages",fontsize=15)
plt.ylabel("Number of users",fontsize=15)
plt.title("List of languages")

plt.barh(x,y,color='r',label="Popularity" )
plt.barh(x,z,color='y',label="Popularity1" )
plt.legend()
plt.show()
```
